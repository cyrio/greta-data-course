# Matplotlib Application Interfaces

```python
%matplotlib inline
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
```

### The Explicit "Axes" interface

In the next sample, each object is explicitly referenced and use to make the next object.

```python
fig = plt.figure() # instantiating an instance of Figure Class
ax = fig.subplots() # create one or more Axes using subplots method on fig we created early
ax.plot([1, 2, 3, 4],[0, 0.5, 1, 0.2]) # call drawing method (plot in this example) on the Axes
```

### The implicit "pyplot" interface
The `pyplot` module shadows most of the `Axes` plotting methods to give the equivalents of above, where the creation of the Figure and Axes is done for the user

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

# Creating , viewing and saving Matplotlib Figures

```python
a=np.array([2, 5, 7, 4, 7, 0, 3, 1, 9, 2])
```

```python
plt.plot(a) # plot the point in the array a
plt.title("My first figure") # Add a title to the figure
plt.xlabel("My x-axis") # Give a label to the x-axis
plt.ylabel("My y-axis") # Give a label to the y-axis
plt.show()
```

```python
fig, ax = plt.subplots() # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]) # Plot some data on the axes.
plt.show()
```
